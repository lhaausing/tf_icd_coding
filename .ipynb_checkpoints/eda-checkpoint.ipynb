```python
from transformers import BertTokenizer
```

```python
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
```

```python
len_sum = []
for i in range(8066):
    text = ';'.join([str(elem) for elem in tokenizer.encode(df1['TEXT'][8065])])
    text = text.split('102')
    text = [sent.split(';') for sent in text]
    lens = [len(elem) for elem in text]
    len_sum.extend(lens)
    if i %100 == 99:
        print("{} rounds.".format(i+1))
```

```python
plt.hist(sorted(len_sum), 300)
```

```python
sum(len_sum)/ len(len_sum)
```

```python
len(len_sum)
```

```python
max(len_sum)
```

```python
#Seperating a sample through [SEP] tokens
text = ';'.join([str(elem) for elem in tokenizer.tokenize(df1['TEXT'][8065])])
text = text.split('[SEP]')
text = [sent.split(';') for sent in text]
lens = [len(elem) for elem in text]
```

```python
for i in range(50):
    print('The {} th line:'.format(i+1))
    print (' '.join(text[i]))
    print('\n')
```
